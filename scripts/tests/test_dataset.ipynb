# Test laion datasets

```python
import sys
sys.path.append("/fsx/home/johnathan/variable-length-embeddings")

from vle.utils import load_config, instantiate_from_config
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2

import torch
torch.manual_seed(2023)

config = load_config("../../configs/training_config_laion.yaml")
dataset = instantiate_from_config(config["data"]).train_dataloader()

batch_idx = 0
stop_iter = 5

for i, batch in enumerate(dataset):
    img = batch[batch_idx].moveaxis(0, -1).numpy()
    img = (img + 1.0) / 2.0 * 255

    plt.figure()
    plt.imshow(img.astype(int))
    
    if i >= stop_iter:
        break
```
